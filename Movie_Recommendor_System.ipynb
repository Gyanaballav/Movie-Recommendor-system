```python
movies = pd.read_csv('/content/drive/MyDrive/Project/Movie Recommendor System/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/drive/MyDrive/Project/Movie Recommendor System/tmdb_5000_credits.csv')
```

```python
#Merging the two dataframe
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
# genres
# id
# keywords
# title
# overview
# Cast
# Crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
import ast
```

```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  counter = 0
  L = []
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter += 1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
df = movies[['movie_id','title','tag']]
```

```python
df['tag'] = df['tag'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tag'] = df['tag'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#To avoid the same kind of words like "love", "loving"
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['tag'] = df['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['tag']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
  movie_index = df[df['title']== movie].index[0]
  distances = similarity[movie_index]
  #sort the distance in the reverse manner
  movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]  #enumerate is used to uphold the position of the movie after sorting
  for i in movies_list:
    print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
